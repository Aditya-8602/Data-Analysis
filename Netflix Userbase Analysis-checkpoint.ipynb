```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Netflix Userbase.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Gender Distribution:
df["Gender"].value_counts().plot(kind='bar', title='Gender Distribution')
```

```python
# Subscription plan distribution
df["Subscription Type"].value_counts().plot(kind='bar', title='Subscription Plan Distribution')
```

```python
# Examine the age distribution of the userbase
age_distribution = df["Age"].describe()
age_distribution
```

```python
# Subscription plan by age group
sns.boxplot(x='Subscription Type', y='Age', data=df)
plt.title('Subscription Plan Popularity by Age')
```

```python
# Assuming a cost per subscription plan
plan_costs = {'Basic': 8.99, 'Standard': 13.99, 'Premium': 17.99}
df["Monthly Revenue"] = df["Subscription Type"].map(plan_costs)

# Total revenue
total_revenue = df["Monthly Revenue"].sum()
print(f'Total Monthly Revenue: ${total_revenue:.2f}')
```

```python
# Analyze monthly revenue based on subscription types
monthly_revenue_analysis = df.groupby("Subscription Type")["Monthly Revenue"].sum()
monthly_revenue_analysis
```

```python
# Determine the geographical distribution of users based on country
country_distribution = df["Country"].value_counts()
country_distribution
```

```python
df["Country"].value_counts().plot(kind='bar', title='User Distribution by Country')
```

```python
# Gender Preference by Age Group:
df['Age_Group'] = pd.cut(df["Age"], bins=[0, 18, 35, 50, 100], labels=['Teen', 'Young Adult', 'Adult', 'Senior'])

df.groupby('Age_Group')["Gender"].value_counts().unstack().plot(kind='bar', stacked=True, title='Genre Preference by Age Group')
```

```python
df["Gender"].value_counts()
```

```python
# Explore device usage among different subscription types
device_usage = df.groupby("Subscription Type")["Device"].value_counts(normalize=True).unstack()
device_usage
```

```python
# Device Usage Analysis:
df["Device"].value_counts().plot(kind='bar', title='Device Usage')
df["Device"].value_counts()
```

```python
# Checking if there's a significant relationship between subscription plan and gender.
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df["Gender"], df["Subscription Type"])

chi2_stat, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2_stat}, P-value: {p}")
```

```python
plt.figure(figsize = (10,8))
sns.pairplot(df[["Age", "Join Date", "Monthly Revenue"]])
plt.show()
```

```python
# Creating age group feature
df["Age_Group"] = pd.cut(df["Age"], bins=[0, 18, 35, 50, 100], labels=['Teen', 'Young Adult', 'Adult', 'Senior'])

print(df["Age_Group"].value_counts())
```

```python
# Monthly User Growth:
df["Join Date"] = pd.to_datetime(df["Join Date"])

df['YearMonth'] = df["Join Date"].dt.to_period('M')

monthly_users = df.groupby('YearMonth').size()

monthly_users.plot(kind='line', title='Monthly User Growth')
plt.xlabel('Year-Month')
plt.ylabel('Number of Users')
plt.show()
```

```python
# Analyze subscription type distribution
subscription_distribution = df["Subscription Type"].value_counts()
subscription_distribution
```

```python
# Identify trends over time based on the join date of users
# Convert join date to datetime format
df["Join Date"] = pd.to_datetime(df["Join Date"], format='%d-%m-%y')
join_trends = df["Join Date"].dt.year.value_counts().sort_index()
join_trends
```

```python
# Examine retention rates by analyzing join date against the last payment date
# Convert the last payment date to datetime format
df["Last Payment Date"] = pd.to_datetime(df["Last Payment Date"], format='%d-%m-%y')
retention_analysis = (df["Last Payment Date"] - df["Join Date"]).dt.days.mean()
retention_analysis
```

```python
# relationship between subscription type vs user age
relationship = df.groupby("Subscription Type")["Age"].mean()
relationship
```

```python
# Classify users based on their gender and age
classification = df.groupby(["Gender", pd.cut(df["Age"], bins=[25, 30, 35, 40, 45, 50])]).size()
classification
```
